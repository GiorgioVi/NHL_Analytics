```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TrophyWinners.csv')
```

```python
MVP = df[df['Award'] == 'MVP']
BESTD = df[df['Award'] == 'BestD']
Goals = df[df['Award'] == 'LeadingGoal']
```

```python
plt.plot(MVP['Year'], MVP['Height'], label='MVP')
plt.plot(BESTD['Year'], BESTD['Height'], label='Norris Winner')
plt.plot(Goals['Year'], Goals['Height'], label='Leading Goal Scorer')


plt.xlabel("Award Year")
plt.ylabel("Height in Centimeters")
plt.title("Historical Height of NHL Award Winners")


plt.legend()
```
